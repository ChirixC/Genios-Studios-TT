```python
import pandas as pd
import re
```

```python
# Variable declarations for later use

file_path = '..\\data\\dataSocialMedia.csv'
clean_file_path = '..\\data\\dataSocialMediaClean.csv'
column_names = ["id", "message", "name", "id_user", "username", "id_post", "link", "date", "user_link", "weekday", "just_emoji", "sentiment", "reply_screen_name", "created_at", "owner", "shortcode", "hour"]
cleaned_lines = []
```

```python
# Read the file as a text file to prevent errors

with open(file_path, 'r', encoding='utf-8') as file: # type: ignore
    lines = file.readlines()[1:]
file.close()
```

```python
# ""(.*?,.*?)"" | RegExp to find all the (,) between ""
# Clear lines of all misplaced symbols

def clean_symbols(line):
    temp_text = re.sub(r'""(.*?,.*?)""', lambda m: m.group(0).replace(',', '$$$'), line)
    temp_text = temp_text.replace(',','\t') # Using TAB (/t) as the original data contains (, and ;) to prevent errors.
    temp_text = temp_text.replace('$$$',',')
    temp_text = temp_text.replace('"','')
    temp_text = temp_text.replace(';','')
    return temp_text

# Clear and append to a list each line
for line in lines:
    cleaned_line = clean_symbols(line)
    cleaned_lines.append(cleaned_line)
```

```python
# Make a new clean file
with open (clean_file_path, 'w', encoding='utf-8') as file:
    file.write('\t'+'\t'.join(column_names)+'\n')
    for line in cleaned_lines:
        file.write(line)
file.close()
```
